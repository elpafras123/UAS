```python
import math
dis = []
for i in range(5000):
    dis.append(math.sqrt((float(databengbeng.iloc[i]['Age'])-30)**2+
                         (float(databengbeng.iloc[i]['Personal Loan'])- 1)**2+
                         (float(databengbeng.iloc[i]['Securities Account'])-0)**2))
                         
```

```python
databengbeng['dis']=dis
databengbeng
```
